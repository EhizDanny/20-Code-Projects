### Sales Prediction and Modelling

```python
# Import Dependencies
from sklearn.model_selection import cross_val_score, GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.feature_selection import chi2
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import math
from sklearn.model_selection import KFold
```

```python
# Drop the non-predictive columns
X = ds.copy() # Copy the dataset
X.drop(['Country', 'City', 'Customer Name', 'Customer ID', 'Order Date', 'Ship Date', 'Postal Code', 'Product Name', 'Product ID'], axis = 1, inplace = True)
y = X.pop('Sales') # Pick out the target column
X.shape
```

```python
# Display Categorical Columns
X.select_dtypes(include=['object']).columns.tolist()
```

```python
# Make a list of the categorical columns
cat_cols = ['Ship Mode', 'Segment', 'State', 'Region', 'Category', 'Sub-Category']
```

```python
# Label Encode Categorical columns
lb = LabelEncoder() # Instantiate the library
for i in cat_cols:
    X[i] = lb.fit_transform(X[[i]])
X.shape
```

```python
# Display new Dataset
X
```

#### Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size = 0.8,  random_state= 2)
print(f'X_train: {X_train.shape}')
print(f'X_test: {X_test.shape}')
print('y_train: {}'.format(y_train.shape))
print('y_test: {}'.format(y_test.shape))
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr_model = dtr.fit(X_train, y_train)
dtr_cv = cross_val_score(dtr, X_train, y_train,cv = KFold(n_splits=5, random_state=1, shuffle=True), scoring = 'r2')
print("%0.2f accuracy with a standard deviation of %0.2f" % (dtr_cv.mean(), dtr_cv.std()))
```
